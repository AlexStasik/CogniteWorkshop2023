```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
from dtwalign import dtw
```

```python
np.random.seed(1234)

n_timepoints = 200+1
t = np.linspace(0, 1, n_timepoints, endpoint=True)

x1 = signal.gausspulse(t-0.5, fc=1.e1) + 0.3*np.random.randn(n_timepoints)
x2 = signal.gausspulse(t-0.3, fc=1.e1) + 0.3*np.random.randn(n_timepoints)

plt.figure()
plt.plot(t, x1)
plt.plot(t, x2)
plt.savefig('unalligned_data.pdf')
plt.show()
```

```python
res = dtw(x1,x2)
```

```python
res.plot_path()
```

```python
plt.figure()
x1_path = res.path[:,0]
x2_path = res.path[:,1]
plt.plot(x1[x1_path],label="aligned query")
plt.plot(x2[x2_path],label="aligned reference")
plt.legend()
plt.ylim(-1.5,1.5)
plt.savefig('ulligned_data.pdf')
plt.show()
```
